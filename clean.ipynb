```python
def random_sample_imputation(df):
    missing_values = df.columns[df.isnull().any()].tolist()
    for val in missing_values:
        # extract a random sample from the column
        random_sample = df[val].dropna().sample(df[val].isnull().sum(), random_state=0)
        # pandas needs to have the same index in order to merge the dataset
        random_sample.index = df[df[val].isnull()].index

        # replace the missing values
        df.loc[df[val].isnull(), val] = random_sample
    return df

df = df2.copy()
dff = random_sample_imputation(df)
dff.head()
```

```python
df3 = pd.read_csv('audible_uncleaned.csv')
df3.head()
```

```python
pattern = r'[^:]*:'
rem = lambda x: re.sub(pattern, '', x)
df3['author'] = df3['author'].apply(rem)
df3['narrator'] = df3['narrator'].apply(rem)
df3['releasedate'] = pd.to_datetime(df3['releasedate'])
df3.head()
```

```python
df3['ratings'] = df3['stars'].str.extract(r'(\d+)\s*ratings').astype(float)
df3['stars'] = df3['stars'].str.extract(r'(\d+(?:\.\d+)?)').astype(float)
df3.head()
```

```python

# pattern = r'(\d+)\s*hrs?|(\d+)\s*mins?'
def convert_to_minutes(time_str):
    hours = 0
    minutes = 0
    # Regex to capture hours and minutes
    hour_match = re.search(r'(\d+)\s*hrs?', time_str)
    minute_match = re.search(r'(\d+)\s*mins?', time_str)
    
    if hour_match:
        hours = int(hour_match.group(1))
    if minute_match:
        minutes = int(minute_match.group(1))
    
    total_minutes = hours * 60 + minutes
    return total_minutes

    
df3['time'] = df3['time'].apply(convert_to_minutes)
df3.head()

```
